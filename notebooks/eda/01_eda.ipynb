```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv(
    r"C:\Projects\EduShield_AI\data\raw\student_dropout_dataset_v3.csv"
)

print("Shape:", df.shape)
df.head()
```

```python
Target Variable Analysis
```

```python
print(df["Dropout"].value_counts())

print("\nPercentage:")

print(
    df["Dropout"].value_counts(normalize=True) * 100
)
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="Dropout"
)

plt.title("Dropout Distribution")

plt.show()
```

```python
missing_values = df.isnull().sum()

missing_values = missing_values[
    missing_values > 0
]

print(missing_values)
```

```python
plt.figure(figsize=(10,5))

missing_values.sort_values().plot(
    kind="barh"
)

plt.title("Missing Values")

plt.show()
```

```python
Numerical Feature Summary
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.select_dtypes(include=["int64","float64"]).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Dropout",
    y="GPA"
)

plt.title("GPA vs Dropout")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Dropout",
    y="Attendance_Rate"
)

plt.title("Attendance vs Dropout")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Dropout",
    y="Stress_Index"
)

plt.title("Stress Index vs Dropout")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="Department",
    hue="Dropout"
)

plt.xticks(rotation=45)

plt.title("Department-wise Dropout")

plt.show()
```
